```python
import random
import matplotlib.pyplot as plt
import numpy as np
col, row = 1000, 1000
N = int(1e6)
```

for $J_{ij} = 1$

```python
for beta in range(-10, 11):
    matrix = np.random.choice([-1,1], size=(col+2,row+2))
    matrix[0,:] = matrix[:,0] = matrix[:,-1] =  matrix[-1,:] = 0
    possible_nodes = []
    for i in range(1,row+1):
        for j in range(1,col+1):
            possible_nodes.append([i,j])
    for simulation in range(N):
        k, j = random.choice(possible_nodes)
        u = np.random.random()
    
        ss = matrix[k][j-1] + matrix[k][j+1] + matrix[k-1][j] + matrix[k+1][j]
        if u<= np.exp(-beta * ss)/ (np.exp(beta * ss)):
            matrix[k][j] = 1
        else:
            matrix[k][j] = -1
    pos = matrix < 1
    plt.imshow(pos)
    plt.title('J = 1,For Beta equal to {}'.format(beta))
    #print('For Beta equal to {}'.format(beta))
    plt.savefig('J = 1,For Beta equal to {}'.format(beta))
```
